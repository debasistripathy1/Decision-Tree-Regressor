```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv("/Users/dev/Desktop/python1/Internshala/Positions_salaries.csv")
```

```python
x=dataset.iloc[:,1].values
x=x.reshape(-1,1)
```

```python
y=dataset.iloc[:,2].values
```

```python
from sklearn.tree import DecisionTreeRegressor as dtr
regressor=dtr(random_state=0)
regressor.fit(x,y)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
```

```python
print(regressor.predict([[16.5]]))
print(regressor.predict([[4.5]]))
print(regressor.predict([[1.5]]))
print(regressor.predict([[6.5]]))
print(regressor.predict([[5.5]]))
```

```python
x_grid=np.arange(min(x),max(x),0.1)
print(x_grid)
x_grid=x_grid.reshape(-1,1)
```

```python
plt.scatter(x,y,color="red")
```

```python
plt.scatter(x,y,color="green")
plt.plot(x_grid,regressor.predict(x_grid),color="red")
plt.show()
```

```python
from sklearn import tree
from sklearn.tree import export_graphviz
export_graphviz(regressor,out_file="./tree.dot")
```
